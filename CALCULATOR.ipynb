```
import tkinter
m = tkinter.Tk()
m.title("Calculator")
m.configure(bg="#282c34")
exp = ""  
text1 = tkinter.StringVar()  

def button_clicked(num):
    global exp
    exp += str(num)
    text1.set(exp)

def compute():
    global exp
    try:
        exp = exp.replace("x", "*")
        result = str(eval(exp))
        text1.set(result)
        exp = result
    except:
        text1.set("Error")
        exp = ""

def remove():
    global exp
    exp = ""
    text1.set(exp)

button_bg = "#4409c1"
button_fg = "white"
op_bg = "#ec616d"
op_fg = "white"
clear_bg = "#47a902"
clear_fg = "white"
entry_bg = "black"
entry_fg = "white"

t1 = tkinter.Entry(
    m, textvariable=text1, font=("Arial", 20), width=15, bg=entry_bg, fg=entry_fg, justify="right"
)
t1.grid(row=0, column=0, columnspan=4)

l1 = tkinter.Button(m, text="1", command=lambda: button_clicked(1), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l2 = tkinter.Button(m, text="2", command=lambda: button_clicked(2), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l3 = tkinter.Button(m, text="3", command=lambda: button_clicked(3), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l4 = tkinter.Button(m, text="4", command=lambda: button_clicked(4), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l5 = tkinter.Button(m, text="5", command=lambda: button_clicked(5), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l6 = tkinter.Button(m, text="6", command=lambda: button_clicked(6), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l7 = tkinter.Button(m, text="7", command=lambda: button_clicked(7), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l8 = tkinter.Button(m, text="8", command=lambda: button_clicked(8), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l9 = tkinter.Button(m, text="9", command=lambda: button_clicked(9), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)
l0 = tkinter.Button(m, text="0", command=lambda: button_clicked(0), width=5, height=2, font=("Arial", 13), bg=button_bg, fg=button_fg)

o1 = tkinter.Button(m, text="+", command=lambda: button_clicked('+'), width=5, height=2, font=("Arial", 13), bg=op_bg, fg=op_fg)
o2 = tkinter.Button(m, text="-", command=lambda: button_clicked('-'), width=5, height=2, font=("Arial", 13), bg=op_bg, fg=op_fg)
o3 = tkinter.Button(m, text="x", command=lambda: button_clicked('x'), width=5, height=2, font=("Arial", 13), bg=op_bg, fg=op_fg)
o4 = tkinter.Button(m, text="/", command=lambda: button_clicked('/'), width=5, height=2, font=("Arial", 13), bg=op_bg, fg=op_fg)
o5 = tkinter.Button(m, text="=", command=compute, width=5, height=2, font=("Arial", 13), bg="#61afef", fg="white")

c0 = tkinter.Button(m, text="AC", command=remove, width=5, height=2, font=("Arial", 13), bg=clear_bg, fg=clear_fg)


l7.grid(row=1, column=0)
l8.grid(row=1, column=1)
l9.grid(row=1, column=2)
o4.grid(row=1, column=3)

l4.grid(row=2, column=0)
l5.grid(row=2, column=1)
l6.grid(row=2, column=2)
o3.grid(row=2, column=3)

l1.grid(row=3, column=0)
l2.grid(row=3, column=1)
l3.grid(row=3, column=2)
o2.grid(row=3, column=3)

c0.grid(row=4, column=0)
l0.grid(row=4, column=1)
o5.grid(row=4, column=2)
o1.grid(row=4, column=3)

m.mainloop()
```
